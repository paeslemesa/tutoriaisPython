## 4. Visualiza√ß√£o dos dados

Agora, vamos gerar alguns gr√°ficos para visualizar nossos dados. Existem v√°rias outras formas, mas vou me basear em detec√ß√£o de dados nulos, histogramas, an√°lise de correla√ß√£o e boxplot.

üåà Antes de mais nada, j√° deixo aqui o [link de op√ß√µes de cores do Matplotlib](https://matplotlib.org/stable/users/explain/colors/colormaps.html) para voc√™s poderem criar gr√°ficos da cor que quiserem!!

```python
# Antes de come√ßar com essas visualiza√ß√µes, vamos tirar a coluna classe do nosso dataframe porque ele √© uma vari√°vel dependente
df_viz = df.drop(columns='classe')

# Conferindo se filtramos certinho
df_viz.head()
```

### 4.1. Detec√ß√£o de dados nulos

Uma an√°lise visual boa de fazer √© criar um mapa de calor e tentar encontrar alguma varia√ß√£o das cores.

No nosso caso, j√° vimos anteriormente que n√£o temos dados nulos. Mas caso tiv√©ssemos, poder√≠amos usar a fun√ß√£o `df.dropna()` ( [documenta√ß√£o dessa fun√ß√£o](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html))

```python
sns.heatmap(df_viz.isna(),cmap = 'Greens')
```

### 4.3. Histogramas

Os histogramas s√£o a quantidade de dados que temos num determinado valor ou conjunto de valor. A partir da an√°lise de histogramas, conseguimos ter uma no√ß√£o se nossos dados segue, por exemplo a [Distribui√ß√£o Normal](https://www.inf.ufsc.br/~andre.zibetti/probabilidade/normal.html).

Vou apresentar aqui dois histogramas: separando os dados por banda e depois separando os dados por classo (mais recomendado).

#### 4.3.1. Histograma dividindo os dados por banda (coluna)

Esse caso √© bem mais simples de gerar o gr√°fico, pois s√≥ precisamos usar `sns.heatmap(df)` que ele automaticamente vai separar os valores pelas colunas.

A gente pode usar a op√ß√£o `kde = True` se quisermos que a linha generalizando os dados apare√ßam. √â quest√£o de opini√£o, eu gosto. Caso queira tirar, basta usar `kde = False`.

```python
fig, ax = (plt.subplots(nrows=1, ncols=1, figsize=(4, 5)))

sns.histplot(df_viz, kde = True)

fig.suptitle("Histograma de cada coluna")
```

#### 4.3.2. Estudando os dados por classe (varia√ß√£o numa √∫nica coluna)

Neste caso, vamos determinar o nosso dataframe inteirinho (incluindo as informa√ß√µes de classe que tiramos antes) e vamos estudar apenas uma banda.

Esta forma de representa√ß√£o j√° nos ajuda com a determina√ß√£o das fei√ß√µes/bandas para usarmos e nos d√° uma dica tamb√©m de como poderemos diferenciar as nossas classes.

```python
fig, ax = (plt.subplots(nrows=1, ncols=1, figsize=(4, 5)))

sns.histplot(data = df, # determinando o dataframe que queremos estudar
             x = "nir", # a coluna que queremos estudar
             hue = "classe", # como vamos separar essa coluna
             kde = True,
             palette=['g', 'b', 'y', 'r'] # a paleta de cores para cada histograma
             )

fig.suptitle("Histograma de cada classe da banda verde")
```

### 4.3. Correla√ß√µes

ü¶ã √â o gr√°fico que eu mais me divirto fazendo! Adoro brincar com as cores e ler esses gr√°ficos!

ü§ñ O gr√°fico de correla√ß√£o vai nos ajudar a entender como as vari√°veis independentes est√£o relacionadas entre si. Sabemos que uma alta correla√ß√£o pode nos atrapalhar para modelos de IA. Mas uma baixa correla√ß√£o pode nos atrapalhar tamb√©m.

üíé Para resumir a leitura do gr√°fico, ele mostra a corrla√ß√£o da `linha i` com a `coluna j`. Ent√£o, quando `i=j`, a correla√ß√£o vai ser sempre 1. A correla√ß√£o varia entre -1 e 1, aonde:
 * 1 √© totalmente correlacionado (mesmo dado)
 * zero √© totalmente n√£o-correlacionado
 * -1 totalmente e inversamente correlacionado (como se fosse o gr√°fico espelhado)

‚õèÔ∏è Ent√£o √© bom usarmos esse gr√°fico para podermos saber se as fei√ß√µes (ou bandas) que estamos utilizando fazem sentido ou se precisamos pensar um pouquinho a mais para escolher as bandas certas para trabalhar. Se por um acaso, voc√™ chegou nesse problema, recomendo tentar dar uma estudada em Minera√ß√£o de Dados (_Data Mining_)

```python
# Gerando a tabela de correla√ß√£o

corr = df_viz.corr()
corr
```

```python
# Para criar um gr√°fico de correla√ß√£o, vamos usar a biblioteca Seaborn novamente <3
fig, ax = (plt.subplots(nrows=1, ncols=1, figsize=(7, 6)))

sns.heatmap(corr, # dataframe com a correla√ß√£o que calculamos ali em cima
            annot =True, # para mostrar os valores  de correla√ß√£o no gr√°fico
            linewidth = 0.5, # para separar as cores com uma linha de espessura 0.5 (recomendo para facilitar visualiza√ß√£o para dalt√¥nicos
            vmin=-1, # valor m√≠nimo da escala de cores
            vmax=1, # valor m√°ximo da escala de cores
            cmap='coolwarm' # paleta de cores utilizada. voc√™ pode selecionar qualquer uma das paletas dispon√≠veis em https://matplotlib.org/stable/tutorials/colors/colormaps.html
            )

fig.suptitle("Correla√ß√£o entre as bandas")
```

## 5. Detec√ß√£o de Outliers com o Pandas üêº

A detec√ß√£o de outliers (pontos fora do padr√£o) √© uma quest√£o bastante questionada na Ci√™ncia de Dados. Nem sempre recomendamos que eles sejam retirados.

Mas, para saber se tira ou n√£o, que tal tentarmos entender que bulhufas √© essa de outlier?


> _O outlier √© um valor que est√° nos pontos extremos de um conjunto de dados (tanto no in√≠cio quanto no fim). Eles podem afetar a observa√ß√£o geral sobre os dados._

image.png

---

Para explicar os outliers, a gente precisa falar de Percentil e Mediana (_median_ em ingl√™s). Vamos selecionar os 25% menores dados (Primeiro Quartil), os 50% menores dados (Segundo Quartil, ou Mediana) e os 75% menores dados (Terceiro Quartil).

Geralmente a observa√ß√£o √© considerada outlier se:
 * ela √© menor que $Q1 - 1.5\cdot(intervalo interquartil)$
 * ela √© maior que $Q3 + 1.5\cdot(intervalo interquartil)$

Para entender melhor, acesse:
  * https://www.escolaedti.com.br/o-que-e-um-box-plot


Aqui, vamos criar o bloxplot apenas para estudarmos a classe Floresta (1).

Para isso, precisaremos filtrar o dataframe usando o "localizador" (`.loc`). Ele funciona mais ou menos assim: localizar no dataframe aonde o dataframe obedece √†s condi√ß√µes que n√≥s estipularmos.

No caso:
`df_viz.loc[df['classe'] == 1]` vai localizar linhas do dataframe __df_viz__ aonde o dataframe original (df) na coluna classe for igual a 1.

```python
fig, ax = plt.subplots(figsize=(10, 3))

sns.boxplot(data=df_viz.loc[df['classe'] == 1],
            orient='h')

fig.suptitle("Boxplot das bandas da classe FLORESTA")
```

### 5.1. O que podemos perceber dos outliers
* A quantidade de outliers √© imensa e podemos perder uma boa quantidade de dados
* Esses pontos podem ser representativos do nosso modelo
